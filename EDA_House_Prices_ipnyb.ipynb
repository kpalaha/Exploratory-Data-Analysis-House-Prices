```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the House Prices train data from GitHub into a pandas DataFrame
df = pd.read_csv('https://github.com/ArinB/MSBA-CA-Data/raw/main/CA01/house-price-train.csv')
```

```python
df = pd.DataFrame(df)
df
```

```python
# print the DataFrame's data types
pd.set_option('display.max_rows', None)
print(df.dtypes)
```

```python
# check for missing values
missing_data = pd.DataFrame(df.isnull().sum(), columns=["Missing Values"])
missing_data
```

```python
sns.histplot(df['SalePrice'])
plt.show()
```

```python
sns.boxplot(x=df['TotalBsmtSF'])
plt.show()
```

```python
# create a boxenplot to visualize the distribution of a numeric column by the categories of a categorical column
sns.boxenplot(x='MSZoning', y='SalePrice', data=df, palette='Set2')
plt.ylabel('SalePrice')
plt.show()
```

```python
sns.scatterplot(x=df['GrLivArea'], y=df['SalePrice'], hue=df['MSZoning'])
plt.show()
```

```python
## Check if the values are unique
unique_values = pd.DataFrame(columns=["Unique Values"])
for row in list(df.columns.values):
  unique_values.loc[row] = [df[row].nunique()]
```

```python
## Check relevatn min or max values
maximum_values = pd.DataFrame(columns=['Maximum Value'])
minimum_values = pd.DataFrame(columns=['Minimum Value'])

for column in df.columns:
    # Exclude non-numeric values before calculating max and min
    numeric_values = pd.to_numeric(df[column], errors='coerce')

    # Calculate and assign maximum value
    maximum_values.loc[column] = [numeric_values.max()]

    # Calculate and assign minimum value
    minimum_values.loc[column] = [numeric_values.min()]
```

```python
## print the DataFrame's data types
data_types = pd.DataFrame(
    df.dtypes,
    columns=['Data Type']
)
data_types
```

```python
#Creating a data quality report
dq_report = data_types.join(missing_data).join(unique_values).join(maximum_values).join(minimum_values)
dq_report
```

```python
## Part 2: Handling Issues in Data Quality Report
df_copy = df.copy()
missing_values_columns = ["LotFrontage", "Alley", "MasVnrType", "MasVnrArea", "BsmtQual", "BsmtCond",
                    "BsmtExposure", "BsmtFinType1", "BsmtFinType2", "Electrical", "FireplaceQu",
                    "GarageType", "GarageYrBlt", "GarageFinish", "GarageQual", "GarageCond"]
```

```python
#Impute Lot Frontage column using median
df_copy['LotFrontage'].fillna(df_copy['LotFrontage'].median(), inplace=True)
```

```python
#Finding out which columns to drop by getting the percentage of missing values
percent_missing = df_copy.isnull().sum() * 100 / len(df_copy)
missing_value_df = pd.DataFrame({'column_name': df.columns,
'percent_missing': percent_missing})
missing_value_df

#In this case, we should delete the columns Alley, PoolQC, Fence, and MiscFeature since their missing values
# percentages are above 60%
```

```python
#Dropping those columns
df_copy = df_copy.drop(["Alley", "PoolQC", "Fence", "MiscFeature"], axis=1)
df_copy
```
